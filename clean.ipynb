```python
import pandas as pd
```

```python
ca_rain = pd.read_csv("Data/Ca_Rainfall_20201103_090000_20210531_235959.csv")
sy_rain = pd.read_csv("Data/Sy_Rainfall_20201103_090000_20210531_235959.csv")
wa_rain = pd.read_csv("Data/Wa_Rainfall_20201103_090000_20210531_235959.csv")
wo_rain = pd.read_csv("Data/Wo_Rainfall_20201103_090000_20210531_235959.csv")
wa_water_level = pd.read_csv("Data/Wa_Creek_Water_Level_20201103_090000_20210531_235959.csv")
```

```python
ca_rain.head(65)
```

```python
dfs = [ca_rain,
           sy_rain,
           wa_rain,
           wo_rain,
           wa_water_level]
df_names = ['ca',
           'sy',
           'wa',
           'wo',
           'wa_water_level']
dfs_clean = []
```

```python
def length_column_name(dfs, df_names):
    for df,name in zip(dfs, df_names):
            df = df.rename(columns={"Unnamed: 0": "datetime",
                                    "9am-9am Rainfall (mm)": f"{name} 9am-9am Rainfall (mm)"
                                  })
            df['datetime'] = df['datetime'].astype('datetime64')
            dfs_clean.append(df)
            print(df.head())
            print(df.shape[0])
```

```python
length_column_name(dfs,df_names)
```

```python
dfs_clean
```

```python
df_concat = pd.concat([dfs_clean[0], dfs_clean[1], dfs_clean[2], dfs_clean[3],dfs_clean[4]], axis=1, join="inner")
df_concat.head(65)
```
